### Find the range of all the columns in the dataset

```python
max_sepal_length = iris['Sepal Length (in cm)'].max()
min_sepal_length = iris['Sepal Length (in cm)'].min()

max_sepal_width = iris['Sepal Width in (cm)'].max()
min_sepal_width = iris['Sepal Width in (cm)'].min()


max_petal_length = iris['Petal length (in cm)'].max()
min_petal_length = iris['Petal length (in cm)'].min()



max_petal_width = iris['Petal width (in cm)'].max()
min_petal_width = iris['Petal width (in cm)'].min()

```

```python
print("Range of Sepal length=",max_sepal_length - min_sepal_length)
print("Range of Sepal Width=",max_sepal_width - min_sepal_width)

print("Range of Petal length=",max_petal_length - min_petal_length)
print("Range of Petal Width=",max_petal_width - min_petal_width)
```
